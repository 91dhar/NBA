```python
import pandas as pd
```

```python
mvps = pd.read_csv("mvps.csv")
```

```python
mvps
```

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]

# selecting only the subset of columns from the mvp dataframe.
```

```python
mvps.head()
```

```python
# cleaning the player data
```

```python
players = pd.read_csv("players.csv")
```

```python
players
```

```python
# deleting some extra columns.

del players["Rk"]
```

```python
players.head()
```

```python
players["Player"].head(50)
```

```python
# remove the asterisk (*) from the player names.

players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players.head(50)
```

```python
players.head(20)

# Greg Anderson has multiple rows, for example. That is because he played for multiple teams in a single season.
# Will be removing multiple rows and club them into 1.
```

```python
# grouping the players together, avoid multiple rows for each player in a specific season.

players.groupby(["Player", "Year"]).get_group(("Greg Anderson", 1991))
```

```python
def single_row(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(20)
```

```python
# no need for multi level index. we can drop the index levels. Running it twice to drop 2 levels.

players.index = players.index.droplevel()
```

```python
players[players["Player"] == "Greg Anderson"]

# this example will make it clear. Greg Anderson now has only 1 row for 1991.
```

```python
# combining the player and mvp data
```

```python
# outer merge. not all of the rows in the player column are in the mvp column. 

combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined.head(5)
```

```python
# now sorting out the players who had MVP votes.

combined[combined["Pts Won"] > 0]
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined
```
